#### INCOME DIFFERENCES FOR WOMEN

The underrespresentation of groups other than men is a continued problem in Data Science. We use the $100000$ USD mark to distinguish high income individuals from the rest.

```python
# All high income brackets (above $100000)
high_income = ['100,000-124,999',
               '125,000-149,999',
               '150,000-199,999',
               '200,000-249,999',
               '250,000-299,999',
               '300,000-500,000',
               '> $500,000']
```

```python
# Add a new column for income above and below $100000
df['HighIncome'] = np.where(df.Q24.isin(high_income), 'Yes', 'No')
```

For the sake of simplicity and to answer our specific research question regarding women, we only extract those that identified as _Man_ or _Woman_.

```python
# Select only binary sex participants
binary_gender = df.loc[(df.Q2 == 'Man') | (df.Q2 == 'Woman')]
```

A contingency table of observed values shows the scale of the problem.

```python
# Contigency table of observed values
pd.crosstab(
    binary_gender.Q2,
    binary_gender.HighIncome,
    margins=True
)
```

```python
# Relative frequency observed value contingency table
pd.crosstab(
    binary_gender.Q2,
    binary_gender.HighIncome,
    margins=True,
    normalize=True
)
```

When selecting only those who identified as either man or women, only $19.7$% were woman. Of these, $4.4$% were in the high income brakcet. For men, the figure as almost double at $8.2$%.

A $\chi^{2}$ test for independence can investigate the dependence between sex and income.

```python
stats.chi2_contingency([[14494, 1295], [3706, 172]])
```

There was a significant association between gender and high income status, $\chi^{2}$ value $63,4$, _p_ value $< 0.01$.

#### UNITED STATES OF AMERICA VS SOUTH AFRICA

We investigate of there is a dependence between country (United States of America and South Africa) and income level (again using the $100000$ USD cut-off.

```python
usa_sa_df = df.loc[df.Q3.isin(['South Africa', 'United States of America'])]
```

```python
pd.crosstab(
    usa_sa_df.Q3,
    usa_sa_df.HighIncome,
    margins=True
)
```

```python
stats.chi2_contingency(
    pd.crosstab(
        usa_sa_df.Q3,
        usa_sa_df.HighIncome
    )
)
```

A total of $2237$ respondents where from the USA compared to only $141$ from South Africa. Thirty-seven percent of the American respondents where in the high income bracket compared to only $5.7$% of the South African respondents, $\chi^{2}$ value $56.2$, _p_ value $< 0.01$.

#### PREDICTING INCOME BRACKET USING MACHINE LEARNING

We generate a random forest machine learning model and use age, sex, country, highest qualification, and whether the respondent uses Python as feature variable in a supervised learning application, predicting income level (high or not).

```python
# Selecting the feature and target variables
variables = [
             'Q1',
             'Q2',
             'Q3',
             'Q4',
             'Q7_Part_1',
             'HighIncome'
]

rf_data = df[variables]
```

We will use the TensorFlow DecisionForest module from Google to create the model.

```python
# Install this package if required
!pip install tensorflow_decision_forests
```

```python
# Importing the required packages
import tensorflow_decision_forests as tfdf
import tensorflow as tf
```

```python
classes = rf_data.HighIncome.unique().tolist() # A list of the two income classes
classes
```

```python
rf_data.HighIncome = rf_data.HighIncome.map(classes.index)
```

The data set is split into a training set, which the decision forest model will use to _learn_ from, and a test set, against whichh we can evaluate the trained machine learning model.

```python
def split(ds, r=0.3):
  test_ind = np.random.rand(len(ds)) < r

  return ds[~test_ind], ds[test_ind]
```

```python
rf_data_train, rf_data_test = split(rf_data)
```

There is class imbalance with $92$% of respondents not in the high income brakcet. Simply predicting _NOT HIGH INCLOME BRACKET_ for all cases, makes you $93$% correct all the time. Our model will have to beat this metric.

```python
rf_data_train.HighIncome.value_counts(normalize=True)
```

The decision forest class is instantiated and the model compiles, prior to passing the training data to the model.

```python
rf_model = tfdf.keras.RandomForestModel()
```

```python
rf_model.compile(
    metrics=['accuracy']
)
```

```python
rf_data_train = tfdf.keras.pd_dataframe_to_tf_dataset(
    rf_data_train,
    label='HighIncome'
)
rf_data_test = tfdf.keras.pd_dataframe_to_tf_dataset(
    rf_data_test,
    label='HighIncome'
)
```

```python
rf_model.fit(
    x=rf_data_train
)
```

The model is now evaluated on data it has never seen (the test data).

```python
evaluation = rf_model.evaluate(
    rf_data_test,
    return_dict=True # Returning a dictionary of metrics
)
```

```python
evaluation.keys() # Metric dictionary keys
```

```python
evaluation.values() # Metric values
```

We see an accuracy in excess of our baseline model.

Below, we plot the decision algorithm learned by this model.

```python
tfdf.model_plotter.plot_model_in_colab(
    rf_model,
    tree_idx=0,
    max_depth=4
)
```

```python
rf_model.make_inspector().variable_importances()
```

We note that the country of origin was the _most important variable_ in the model.
