Import the numpy and the pandas packages using the namespace abbreviations `np` and `pd`.

Run the code cell below to import the express module from plotly and set the them for plot.

```python
import plotly.express as px
import plotly.io as pio

pio.templates.default = 'plotly_white'
```

Run the code cell below that generates random values for a variable in a population.

```python
np.random.seed(42)
pop_vals = np.random.normal(loc=300,
                            scale=50,
                            size=2000)
```

Select a random sample of $100$ values from the population and assign the array to the computer variable `my_sample`. Set the random seed value to $1200$.

```python
# Enter your solution here
np.random.seed(1200)
my_sample = np.random.choice(pop_vals, 100)
```

Display a histogram of the sample values. Add an appropriate title and _x_ axis label.

```python
px.histogram(
    x=my_sample,
    title='Histogram of variable values',
    labels={'x':'Variable values'}
)
```

Calcualate the mean and standard deviation parameters for the population.

```python
# Enter your solution here
pop_vals.mean(), pop_vals.std(ddof=0)
```

Calculate the mean and the standard deviation for the sample.

```python
# Enter your solution here
my_sample.mean(), my_sample.std(ddof=1)
```

Run the code cell below that generates $1000$ means from repeated sampling of $100$ subjects from the population.

```python
means_1000 = [np.mean(np.random.choice(pop_vals, 100)) for i in range(1000)]
```

What is the mean of this sampling distribution of means?

```python
# Enter your solution here
np.mean(means_1000)
```
