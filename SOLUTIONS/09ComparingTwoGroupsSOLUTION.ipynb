The test statistic of this study is the difference in mean height between the two groups. Run the code below that calculate the difference in mean heights.

```python
stat = np.mean(df.loc[df.Group == 'I'].Height) - np.mean(df.loc[df.Group == 'II'].Height)
stat
```

Under the null hypothesis there is no difference in mean height between the two groups. The code below reassigns each subejct to a random group and generates $1000$ sampling mean differences.

```python
mean_stat = []
height_copy = height.copy()

for i in range(1000):
  np.random.shuffle(height_copy)
  mean_stat.append(np.mean(height_copy[0:127]) - np.mean(height_copy[128:200]))
```

Run the code cell below to show a histogram of all the test statistics.

```python
px.histogram(x=mean_stat,
             title='Test statistic sampling distribution',
             labels={'x':'Differences in means'})
```

What proportion of the test statistic sampling distribution values are less than the mean from your study?

```python
# Enter your solution here
np.sum(np.array(mean_stat) < stat) / 1000
```

What proprtion of the test statistic sampling distribution values are more than the absolute value of your test statistic?

```python
# Enter your solution here
np.sum(np.array(mean_stat) > -stat) / 1000
```

What was the likelihood of your test statistic (value or more extreme)?

```python
# Enter your solution here
(np.sum(np.array(mean_stat) < stat) / 1000) + (np.sum(np.array(mean_stat) > -stat) / 1000)
```

How does this compare to the _p_ value using Student's _t_ test below.

```python
stats.ttest_ind(df.loc[df.Group == 'I'].Height,
                df.loc[df.Group == 'II'].Height)
```
