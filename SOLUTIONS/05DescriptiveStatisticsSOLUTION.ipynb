Show the average value for the `savings` variable for each of the groups in the `home_loan` variable.

What are the median savings of those $50$ and over with no children?

```python
# Enter your solution here
df.loc[(df.age >= 50) & df.children == 0, 'savings'].median()
```

What is the median savings of those $50$ and over with chidren?

```python
# Enter your solution here
df.loc[(df.age >= 50) & (df.children != 0), 'savings'].median()
```

Calculate the variance in savings for all the sample sapce elements of the `invest` column for those younger than $50$.

```python
# Enter your solution here
df.loc[df.age < 50].groupby('invest')['savings'].var()
```

Calculate the quartile values of savings for each of the home loan groups.
