```python
import plotly.express as px
import plotly.io as pio

pio.templates.default = 'plotly_white'
```

```python
np.random.seed(28)
fracture_strength = np.random.normal(loc=30, scale=7, size=80) # In MPa
```

Calculate the mean and standard deviation of the fracture strength for this sample of $80$ subjects.

```python
# Enter your solution here
np.mean(fracture_strength), np.std(fracture_strength, ddof=1)
```

Assign the mean for your study as `my_mean`.

```python
# Enter your solution here
my_mean = np.mean(fracture_strength)
```

Assign the number of subjects in your sample to the computer variable `size`. Hint: Use the `len` function.

```python
# Enter your solution here
size = len(fracture_strength)
size
```

Create a Python list object called `means` and assign to $5000$ mean values using bootstrap resampling from the sample. Hint: Use list comprehension. remember to use the same sample size and use replacement.

```python
# Enter your solution here
means = [np.mean(np.random.choice(fracture_strength, size=size, replace=True)) for i in range(5000)]
```

Run the code cell below to generate a histogram of the bootstrap resamples means.

```python
px.histogram(
    x=means,
    title='Bootstrap resampled Means',
    labels={'x':'Means'}
)
```

There are $5000$ values in the `means` array. Run the code cell below to creare a sorted array of these values.

```python
means_sort = np.sort(means)
```

Now that all the bootstrapped means are in ascending order, save the indices that represents the $2.5$ and the $97.5$ percentile values as the computer varaibles `k_2_5` and `k_97_5`. Hint: Pass the calculation as an argument to the `int` function.

```python
# Enter your solution here
k_2_5 = int(2.5 / 100 * 5000)
k_97_5 = int(97.5 / 100 * 5000)
```

Calculate the lower and upper limits of the $95$% confidence intervals for your study mean. Assign the values to the computer variable `lower` and `upper`.

```python
# Enter your solution here
lower = means_sort[k_2_5]
upper = means_sort[k_97_5]
```

Run the code cell below to see if the $95$% confidence interval values make sense with respect to the distribution of bootstrapped means and the actual mean for your study.

```python
import plotly.figure_factory as ff
import plotly.graph_objects as go

height = 0.4

ff.create_distplot(
    [means],
    ['Resampled means'],
    curve_type='normal'
).add_trace(
    go.Scatter(
        x=[lower, lower],
        y=[0, height],
        name='2.5%',
        mode='lines',
        marker=dict({'color':'orange'})
    )
).add_trace(
    go.Scatter(
        x=[upper, upper],
        y=[0, height],
        name='97.5%',
        mode='lines',
        marker=dict({'color':'orange'})
    )
).add_trace(
    go.Scatter(
        x=[my_mean, my_mean],
        y=[0, height],
        name='Sample mean',
        mode='lines',
        marker=dict({'color':'green'})
    )
).update_layout(
    title='Bootstrap resampled means',
    xaxis=dict(title='Bootstrapped means'),
    yaxis=dict({'title':'Density'})
)
```
