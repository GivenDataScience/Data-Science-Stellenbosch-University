Import the `crops.csv` file and assign it to the computer variable df.

```python
# Enter your solution here
df = pd.read_csv('crops.csv')
```

Display the first five rows of data.

```python
# Enter your solution here
df.head()
```

Use the `info` method to display the column names, number of non-null values, and the data type of each variable.

```python
# Enter your solution here
df.info()
```

Add a column called `Cost` that adds the `CultivationCost` and the `ProductionCost`.

```python
# Enter your solution here
df['Cost'] = df.CultivationCost + df.ProductionCost
```

Calculate the frequency of each sample space element for the `Crop` variable.

```python
# Enter your solution here
df.Crop.value_counts()
```

Create a computer variable called `totals` to hold a list of the frequencies above.

```python
# Enter your solution here
totals = df.Crop.value_counts().values.tolist()
```

Create a variable called `levels` to hold all the crop types.

```python
# Enter your solution here
levels = df.Crop.value_counts().index.tolist()
```

Create a bar plot to visualise the number of instances of each of the crop types in the data set. Add an appropriate title and axes labels.

```python
# Enter your solution here
crop_fig = go.Figure()

crop_fig.add_trace(
    go.Bar(
        x=levels,
        y=totals
    )
).update_layout(
    title='Count fo each crop type in the data set',
    xaxis={'title':'Crop type'},
    yaxis=dict(title='Count')
)
```

Create a histogram (using the graph_objects module) of the `Cost` variable (created above). Add an appropriate title and axes labels.

```python
# Enter your solution here
go.Figure(
    go.Histogram(
        x=df.Cost
    )
).update_layout(
    title='Total Cost',
    xaxis={'title':'Cost'},
    yaxis={'title':'Frequency'}
)
```

Create the same histogram as above, but set the number of bins to $14$ and the opacity to $0.75$. Hint: Read the documentation at https://plotly.com/python/histograms/

```python
# Enter your solution here
go.Figure(
    go.Histogram(
        x=df.Cost,
        nbinsx=14
    )
).update_layout(
    title='Total Cost',
    xaxis={'title':'Cost'},
    yaxis={'title':'Frequency'}
).update_traces(opacity=0.75)
```

Create a box-and-whisker plot (using the express module) of the `ProductionCost` variable for each of the `Crop` sample space elememts. Add a title and improve on the _y_ axis label.

```python
# Enter your solution here
px.box(
    df,
    x='Crop',
    y='ProductionCost',
    title='Production Cost Per Crop Type',
    labels={'ProductionCost':'Production Cost'}
)
```

Create a scatter plot of `ProductionCost` versus `CultivationCost` using the express module. Add an appropriate title and improive the axes labels.

```python
# Enter your solution here
px.scatter(
    df,
    x='ProductionCost',
    y='CultivationCost',
    title='Cultivation cost vs production cost',
    labels={'ProductionCost':'Production cost',
            'CultivationCost':'Cultivation cost'}
)
```

Recreate the plot above, but separate the data by the `Crop` variable. Display a rug plot for the `ProductionCost` variable and a histogram for the `CultivationCost` variable.

```python
# Enter your solution here
px.scatter(
    df,
    x='ProductionCost',
    y='CultivationCost',
    color='Crop',
    title='Cultivation cost vs production cost',
    labels={'ProductionCost':'Production cost',
            'CultivationCost':'Cultivation cost'},
    marginal_x='rug',
    marginal_y='histogram'
)
```
